```python
def funct():
    base_url='https://www.timesjobs.com/candidate/job-search.html?from=submit&searchType=Home_Search&luceneResultSize=25&postWeek=60&cboPresFuncArea=35&pDate=Y&sequence='
    rest_url='&startPage=1'
    page_number=[]          # all url page append in page_number.
    for number in range(1,41):    # looping the numbers  in range.
        page_number.append(base_url+str(number)+rest_url)
    import requests                                  # requests function used for read html code.
    from bs4 import BeautifulSoup                    # BeautifulSoup function used for convert string to lxml format .
    import pandas as pd
    total_details=[]        # all value in dictionary append in total_details .
    for url in page_number:
        first_page_url_data=requests.get(url).text  
        covert_in_lxml=BeautifulSoup(first_page_url_data,'lxml')
        first_page_common_class=covert_in_lxml.find_all('li',class_='clearfix job-bx wht-shd-bx')
        for item in first_page_common_class:
            productname=item.find('a').text
            space_removing=productname.strip()              # removeing space from both end .
            centering_name=space_removing.center(50)        # centerizing the productname .
            
            company_name=item.find('h3',class_='joblist-comp-name').text
            company_name1=company_name.replace('(More Jobs)','')             # replacing '(More Jobs)' , 'blank'
            cleaned_company=company_name1.strip()
            name_centering=cleaned_company.center(50)
            
            Experience=item.find('li').text
            cleaning_Experience=Experience[11:-1]              #by index number Experience selected 
            cleaned_Experience=cleaning_Experience.center(50)
            
            location=item.find('span').text
            space_removing_location=location.strip()
            replaced_location=space_removing_location.replace("(More Jobs)","")
            cleaned_location=replaced_location.center(50)
            
            job_description=item.find('ul',class_="list-job-dtl clearfix")
            getting_job_description=job_description.find('li').text     # selecting getting_job_description from job_description
            replacing=getting_job_description.replace("More Details",".") # replacing "More Details" by " . " 
            giving_range=replacing[18:-1]                           #by index number job_description selected .
            cleaned_job_description=giving_range.strip()
            
            keyskills=item.find('span',class_='srp-skills').text
            cleaning_keyskills=keyskills.strip()
            cleaned_keyskills=cleaning_keyskills.center(50)
            
            
            link=item.find('a')['href'] # getting url link by indexing ['href']
            
            all_details_in_dictionary={                      # adding all values in dictionary .
                'Product_Name':centering_name,
                'Company_Name':name_centering,
                'Experience':cleaned_Experience,
                'Location':cleaned_location,
                'Job_Description':cleaned_job_description,
                'Keyskills':cleaned_keyskills,
                'Links_of_pages':link
                
            }
            total_details.append(all_details_in_dictionary) #adding all_details_in_dictionary in list 
    
    data_framing=pd.DataFrame(total_details)    # by pandas function importing dataFrame 
    return(data_framing)             # gives output in tabeleau form 
    data_framing.to_excel('my_project1.xlsx')                      #  extracting output in excel format by  by using 'xlsx.format'
funct()
```
